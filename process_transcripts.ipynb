```python
# import LM
path = "Loughran-McDonald_MasterDictionary_1993-2021.csv"

lm_df = pd.read_csv(path)

lm_df
```

```python
lm = {}

for i, row in lm_df.iterrows():
    # print(row)
    word = str(row['Word']).lower()
    if row['Negative'] != 0:
        lm[word] = -1
    if row['Positive'] != 0:
        lm[word] = 1
    
lm
```

```python
def sentscore(tokens):
    # tokens = process(string)
    total = 0
    for token in tokens:
        if token in lm:
            total += lm[token]
            # print(token)
            # print(lm[token])
    # print(total)
    # print(tokens)
    return total
```

```python
df['tokens'] = df['condensed'].apply(lambda x : process(x, stem=False))
```

```python
df['numtokens'] = df['tokens'].apply(len)
```

```python
df['lm_sent'] = df['tokens'].apply(sentscore)
```

```python
def get_lname(string):
    split = string.split(' ')
    if len(split) == 1:
        return None
    else:
        return split[-1][:-1].lower()
```

```python
df['lname'] = df['speaker'].map(get_lname)
```
